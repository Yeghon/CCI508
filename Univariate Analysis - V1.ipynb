```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv("nairobi_schools.csv")
X = data.C
Y = data.E
plotFunction(X, Y)
```

```python
def plotFunction(X, Y):
    plt.figure(figsize=(15, 7))
    plt.xlim([min(X)-0.2, max(X)+0.2])
    plt.ylim([min(Y)-0.05, max(Y)+0.05])
    plt.scatter(X,Y)
    plt.title("Scatter Plot for the dataset")
    plt.xlabel("X")
    plt.ylabel("Y")
```

```python
# the function of h(x) = mx + c, m is the gradient while c is the y intercept
theta1 = 0
theta0 = 0
L = 0.0015 # the learning rate
iterations = 10000
xTr = np.asarray(X[:1024]) #x training data
yTr = np.asarray(Y[:1024]) #y training dataset
xTs = np.asarray(X[1024:]) #x testing data
yTs = np.asarray(Y[1024:]) #y testing dataset
elems = float(len(xTr)) # the number of elements in X
```

```python
histTheta0 = []
histTheta1 = []
for i in range(iterations):
    Y_pred = theta1*X + theta0
    D_theta1 = (-2/elems) * sum(X * (Y-Y_pred))
    D_theta0 = (-2/elems) * sum(Y -Y_pred)
    
    theta1 = theta1 - L * D_theta1
    theta0 = theta0 - L * D_theta0
    
    histTheta0.append(theta0)
    histTheta1.append(theta1)
    

print(theta1,theta0)
```

```python
plt.figure(figsize=(15, 5))
plt.scatter(histTheta0, histTheta1)
# histTheta0[:10]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(histTheta0, histTheta1)
```

```python
# Scatter test dataset and plot linear line of regression
plotFunction(xTs, yTs)
```

```python
def generateY(x):
    return theta1 + theta0 * x
```

```python
# Predicted line of linear regression
plt.figure(figsize=(10, 8))
plt.scatter(xTs, yTs)
newYvalues = list(map(generateY, xTs))
plt.plot(xTs, newYvalues)
```

```python
# Building the model
m = 0.0
c = 0.0

L = 0.0015  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

theta0hist = []
theta1hist = []

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
    theta0hist.append(m)
    theta1hist.append(c)
    
    
    
print (m, c)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(theta1hist, theta0hist)
```
