```python
import nltk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokenized = word_tokenize(song)
tokenized
```

```python
def plotFreq(words, size):
    wordDist = nltk.FreqDist(words)
    plt.figure(figsize=(15, 4))
    wordDist.plot(size)
```

```python
plotFreq(tokenized)
```

```python
''' remove punctuations '''
import string
un_punctuated = []
for word in tokenized:
    un_punctuated.append(word.translate(str.maketrans('','',string.punctuation)))
un_punctuated
```

```python
plotFreq(un_punctuated)
```

```python
''' remove white spaces '''
no_spaces = []
for x in un_punctuated:
    if len(x) >=1:
        no_spaces.append(x)
no_spaces
```
