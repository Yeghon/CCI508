```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/Wikipedia:List_of_English_contractions'
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
table = soup.find_all('table', attrs={'class':'wikitable sortable'})
```

```python
headers = []
heads = soup.find_all('th')
for x in heads:
    headers.append(x.text)
headers
```

```python
data = []
tableData = soup.find_all('tr')
for cell in tableData:
    dataRow = []
    td = soup.find_all('td')
    for x in td:
        dataRow.append(x)
    data.append(dataRow)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(data)
```
