```python
def scatterPlot(x, y):
    plt.figure(figsize=(15, 4))
    plt.scatter(x, y)
```

```python
def plotValues(x, y):
    plt.figure(figsize=(15, 4))
    plt.plot(x, y)
```

```python
''' data normalization, 
option 1. scaling..... xnew = (x_old - x_mean) / x_std
option 2. normalization .... x_new = (x - x_min) / (x_max - x_min'''

def scaleValues(data):
    datamean = np.mean(data)
    datastd = np.std(data)
    newData = []
    for x in data:
        newData.append((x - datamean) / datastd)
    return newData

def normalizeData(data):
    normalizedData = []
    for x in data:
        normalizedData.append((x - np.min(data)) / (np.max(data) - np.min(data)))
    return normalizedData

```

```python
''' filter by Nairobi County'''
ncSchools = df.loc[df.County == 'NAIROBI']
ncSchools.head()
```

```python
'''filter by no of classrooms and total enrollment'''

X = ncSchools.iloc[:, 10]
y = ncSchools.iloc[:, 17]
```

```python
X
```

```python
''' remove zero values '''
X = X.loc[X > 0]
X
```

```python
y = y.loc[y > 0]
```

```python
X_scaled = []
mean = np.mean(X)
std = np.std(X)
for x in X:
    X_scaled.append((x - mean) / std)
    
```

```python
X_scaled
```

```python
y_scaled = []
mean = np.mean(y)
std = np.std(y)
for x in y:
    y_scaled.append((x - mean) / std)
   
```

```python
y_scaled
```

```python
df = pd.DataFrame({"X":X_scaled, "y":y_scaled})
df.head()
```

```python
X = df.iloc[:, 0]
y = df.iloc[:, 1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10, 4))
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(X, y)
plt.show()
```

```python
''' building the model '''
m = 0
c = 0

L = 0.0015
epochs = 10000

n = float(len(X))

m_ = []
c_ = []
''' gradient descent '''
for i in range(epochs):
    Y_pred = m * X + c
    D_m = (-2/n) * sum(X * (y - Y_pred))
    D_c = (-2/n) * sum(y - Y_pred)
    m = m - L * D_m
    m_.append(m)
    c = c - L * D_c
    c_.append(c)
    
```

```python
plt.figure(figsize=(15, 4))
plt.plot(m_, c_)
plt.show()
```

```python
print(m, c)
```

```python
from scipy import stats
```

```python
r = stats.linregress(X, y)
r
```

```python
''' predicted y '''
ypredicted = []
for x in X:
    ypredicted.append(m * x + c)
    
ypredicted
```

```python
plt.figure(figsize=(15, 4))
plt.scatter(X, y)
plt.plot(X, ypredicted, color='yellow')
```

```python
ydf = pd.DataFrame({"X":X, "Actual_Y":y, "Predicted_Y":ypredicted})
ydf.head()
```

```python
y_predicted_ = []
for x in X:
    y_predicted_.append(r.slope * x + r.intercept)
    
y_predicted_
```

```python
ydf = pd.DataFrame({"X":X, "Actual_Y":y, "My_predicted_Y":ypredicted, "Machine_predicted_Y":y_predicted_})
ydf.head()
```

```python
plt.figure(figsize=(15, 4))
plt.scatter(ydf.X, ydf.Actual_Y)
plt.plot(ydf.X, ydf.My_predicted_Y, color='yellow')
plt.scatter(ydf.X, ydf.Machine_predicted_Y, color='red')
```
