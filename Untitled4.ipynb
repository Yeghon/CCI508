```python
from nltk.tokenize import word_tokenize
```

```python
tokenized =  word_tokenize(song)
tokenized
```

```python
len(tokenized)
```

```python
import nltk
freqdist = nltk.FreqDist(tokenized)
freqdist.plot(10)
```

```python
no_puncts = []
import string
for x in tokenized:
    no_puncts.append(x.translate(str.maketrans('','',string.punctuation)))
no_puncts
```

```python
freqdist = nltk.FreqDist(no_puncts)
freqdist.plot(10)
```

```python
no_whitespace = []
for x in no_puncts:
    if len(x) is not 0:
        no_whitespace.append(x)
        pass
    pass
no_whitespace
```

```python
freqdist = nltk.FreqDist(no_whitespace)
freqdist.plot(10)
```

```python
from nltk.corpus import stopwords
```

```python
stops = stopwords.words('english')
print(stops)
```

```python
no_stops = []
for x in no_whitespace:
    if x.lower() not in stops:
        no_stops.append(x.lower())
        pass
    pass
no_stops
```

```python
freqdist = nltk.FreqDist(no_stops)
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 4))
freqdist.plot()
```
